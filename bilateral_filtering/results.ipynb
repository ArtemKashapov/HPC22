```python
import matplotlib.pyplot as plt
```

```python
timings0 = [38.80246966666667, 5.6701162, 0.034379433333333376]
timings1 = [7.193564266666667, 0.9947659333333334, 0.007097100000000041]
timings2 = [2.829615, 0.4099867, 0.0032579333333332774]
```
